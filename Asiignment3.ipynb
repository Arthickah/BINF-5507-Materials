```python
#Assignment3
#DSBCAN Algoritham overview 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans, AgglomerativeClustering



moons, _ = make_moons(n_samples=300, noise=0.05, random_state=42)
blobs, _ = make_blobs(n_samples=300, centers=3, cluster_std=[1.0, 2.5, 0.5], random_state=42)


models = [
    ("DBSCAN (Moons)", DBSCAN(eps=0.3, min_samples=5).fit_predict(moons)),
    ("K-Means (Moons)", KMeans(n_clusters=2).fit_predict(moons)),
    ("Hierarchical (Moons)", AgglomerativeClustering(n_clusters=2).fit_predict(moons)),
    ("DBSCAN (Blobs)", DBSCAN(eps=1.0, min_samples=10).fit_predict(blobs)),
    ("K-Means (Blobs)", KMeans(n_clusters=3).fit_predict(blobs)),
    ("Hierarchical (Blobs)", AgglomerativeClustering(n_clusters=3).fit_predict(blobs))
]


plt.figure(figsize=(15, 8))

for i, (title, labels) in enumerate(models, 1):
    plt.subplot(2, 3, i)
    plt.scatter(moons[:,0], moons[:,1], c=labels, cmap='viridis') if i <= 3 else plt.scatter(blobs[:,0], blobs[:,1], c=labels, cmap='viridis')
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()
```
